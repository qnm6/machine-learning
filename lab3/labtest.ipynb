```python
import pandas as pd

# Load the CSV with no header
df = pd.read_csv(r"C:\Users\mshms\Documents\car+evaluation\car.data", header=None)

# Manually set column names
df.columns = [
    "buying", "maint", "doors", "persons",
    "lug_boot", "safety", "class"
]

# Convert numeric-like columns
df['doors'] = df['doors'].replace({'5more': 5}).astype(int)
df['persons'] = df['persons'].replace({'more': 5}).astype(int)

# Check columns and dtypes
print(df.columns)
print(df.dtypes)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['persons'], bins=5)
plt.title("Distribution of Number of Persons")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['doors'], bins=5)
plt.title("Distribution of Number of Doors")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by 'class' and count
class_counts = df.groupby('class').size()

# Plot
plt.figure(figsize=(10,5))
class_counts.plot(kind='bar')
plt.title("Number of Cars per Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
safety_counts = df.groupby('safety').size()

plt.figure(figsize=(10,5))

plt.ylim(0, 700)
plt.xlim(0,700)
safety_counts.plot(kind='bar', color='orange')
plt.title("Number of Cars per Safety Level")
plt.xlabel("Safety Level")
plt.ylabel("Count")
plt.show()
```

```python
buying_counts = df.groupby('buying').size()

plt.figure(figsize=(10,5))
buying_counts.plot(kind='bar', color='green')
plt.title("Number of Cars per Buying Price")
plt.xlabel("Buying Price")
plt.ylabel("Count")
plt.show()
```
